### ChatGPT 代码测试

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(data_path + "smoking.csv")

import seaborn as sns

sns.countplot(x="smoking", data=df, palette="dark")
plt.show()

sns.catplot(x="smoking", hue="gender", col="age",
            data=df, kind="count",
            height=4, aspect=.7, palette="dark")
plt.show()

sns.relplot(x="age", y="height(cm)", hue="smoking",
            data=df, palette="dark")
plt.show()
```
