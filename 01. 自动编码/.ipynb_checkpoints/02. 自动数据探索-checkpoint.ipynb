### ChatGPT 代码测试

```python
import pandas as pd
import matplotlib.pyplot as plt

# 加载数据集
df = pd.read_csv(data_path + "user_churn.csv")

# 查看数据的前几行
print(df.head())

# 绘制用户流失比例的柱状图
churn_count = df["Churn"].value_counts()
plt.bar(churn_count.index, churn_count.values)
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.title("User Churn Distribution")
plt.show()

# 绘制各个特征与用户流失的关系
for col in df.columns[:-1]:
    pivot = df.pivot_table(index=col, values="Churn", aggfunc="mean")
    pivot.plot(kind="bar")
    plt.xlabel(col)
    plt.ylabel("Churn Rate")
    plt.title(f"Churn rate by {col}")
    plt.show()
```
