### ChatGPT 代码测试

```python
import random
import csv

# 随机生成卡号
def random_card_number():
    return "".join(str(random.randint(0, 9)) for _ in range(16))

# 随机生成信用额度
def random_credit_limit():
    return random.randint(1000, 100000)

# 随机生成是否逾期
def random_is_overdue():
    return random.choice(["是", "否"])

# 随机生成持卡时长
def random_card_holding_duration():
    return random.randint(1, 20)

# 生成随机假数据
def generate_fake_data():
    data = []
    for i in range(1000):
        data.append([
            random_card_number(),
            random_credit_limit(),
            random_is_overdue(),
            random_card_holding_duration()
        ])
    return data

# 将随机数据写入 CSV 文件
def write_to_csv(data):
    with open(result_path + "fake_data.csv", "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["卡号", "信用额度", "是否逾期", "持卡时长"])
        writer.writerows(data)

# 生成并写入随机假数据
if __name__ == "__main__":
    data = generate_fake_data()
    write_to_csv(data)
```
